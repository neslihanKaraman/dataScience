# Birds data set

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.patches as mpatches
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
base_url=r"data\birds.csv"
birds  = pd.read_csv(base_url)
birds.head()
```

```python
birds.info()
```

```python
family_counts = birds["Family"].value_counts()
print(family_counts)
print(type(family_counts))
```

```python
family_counts = birds["Family"].value_counts()
# Seri -> Dataframe
family_counts_df = pd.DataFrame({'Family': family_counts.index, 'SpeciesCount': family_counts.values})
family_counts_df
```

```python
family_counts = birds["Family"].value_counts()
# Seri -> Dataframe
family_counts_df = pd.DataFrame({'Family': family_counts.index, 'SpeciesCount': family_counts.values})

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='Family', y='SpeciesCount', data=family_counts_df)
plt.xticks(rotation=90)
plt.xlabel('Family')
plt.ylabel('Number of Species')
plt.title('Number of Species per Family')
plt.show()
```

```python
family_counts = birds["Family"].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=family_counts.index, y=family_counts.values)
plt.xticks(rotation=90)
plt.xlabel('Family')
plt.ylabel('Number of Species')
plt.title('Number of Species per Family')
plt.show()
```
