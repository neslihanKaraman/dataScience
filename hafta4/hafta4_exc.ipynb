# All Fruit Data

```python
import requests
import pandas as pd
```

```python
URL = 'https://www.fruityvice.com/api/fruit/all'
res = requests.get(URL)
fruit_data = res.json()
fruit_data
```

```python
fruit_data[0]['nutritions']
```

```python
fruit_data[0]['nutritions']['calories']
```

```python
for fruit in fruit_data:
    if fruit.get('family') == 'Rosaceae':
        print(fruit['name'])
```

```python
selected_family = []
for fruit in fruit_data:
    if fruit.get('family') == 'Rosaceae':
        selected_family.append(fruit['nutritions'])
selected_family
```

```python
# Nutritional values of fruits from the family Resaceae
def get_filtered_family_data():
    df = pd.DataFrame(columns=['family', 'name','calories','fat','sugar','carbohydrates','protein']) 
    for fruit in fruit_data:
        if fruit.get('family') == 'Rosaceae':
             df = df._append({
                'family': fruit.get('family'),
                'name': fruit.get('name'),
                'calories': fruit.get('nutritions', {}).get('calories'),
                'fat': fruit.get('nutritions', {}).get('fat'),
                'sugar': fruit.get('nutritions', {}).get('sugar'),
                'carbohydrates': fruit.get('nutritions', {}).get('carbohydrates'),
                'protein': fruit.get('nutritions', {}).get('protein')
        }, ignore_index=True)
    return df

filtered_fruit_df = get_filtered_family_data()
print(filtered_fruit_df)
```

```python
for fruit in fruit_data:
    if fruit.get('genus') == 'Citrus':
        print(fruit['name'])
```

```python
selected_genus = []
for fruit in fruit_data:
    if fruit.get('genus') == 'Citrus':
        selected_genus.append(fruit['nutritions'])
selected_genus
```

```python
# Nutritional values of fruits from the genus Citrus and calories <40
def get_filtered_genus_data():
    df = pd.DataFrame(columns=['genus','name', 'calories','fat','sugar','carbohydrates','protein']) 
    for fruit in fruit_data:
        if fruit.get('genus') == 'Citrus'and fruit.get('nutritions', {}).get('calories', 0) < 40:
             df = df._append({
                'genus': fruit.get('genus'),
                'name': fruit.get('name'),
                'calories': fruit.get('nutritions', {}).get('calories'),
                'fat': fruit.get('nutritions', {}).get('fat'),
                'sugar': fruit.get('nutritions', {}).get('sugar'),
                'carbohydrates': fruit.get('nutritions', {}).get('carbohydrates'),
                'protein': fruit.get('nutritions', {}).get('protein')
        }, ignore_index=True)
    return df

filtered_fruit_df = get_filtered_genus_data()
print(filtered_fruit_df)
```

```python
selected_order = []
for fruit in fruit_data:
    if fruit.get('order') == 'Ericales':
        selected_order.append(fruit['name'])
selected_order
```

```python
# order,family,calories values of fruits from the order Ericales
def get_filtered_order_data():
    df = pd.DataFrame(columns=['order','name','family', 'genus','calories']) 
    for fruit in fruit_data:
        if fruit.get('order') == 'Ericales':
             df = df._append({
                'order': fruit.get('order'),
                'name': fruit.get('name'),
                'family': fruit.get('family'),
                'genus': fruit.get('genus'),
                'calories': fruit.get('nutritions', {}).get('calories')
        }, ignore_index=True)
    return df

filtered_fruit_df = get_filtered_order_data()
print(filtered_fruit_df)
```
